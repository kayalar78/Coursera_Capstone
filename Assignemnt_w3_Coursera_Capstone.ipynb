# Importing the Latitude/Longitude Data

Let's read the data in.

```
from google.colab import files
latlong = files.upload()
```

Let's convert the data into a data frame and name it as "latlong"

```
import io
latlong = pd.read_csv(io.BytesIO(latlong['Geospatial_Coordinates.csv']))
# Dataset is now stored in a Pandas Dataframe
```

```
latlong.shape
```

```
latlong.head()
```

Now, we will join the wikipedia table with the latlong table so that we have alt long data for each row.


```
tablewithlatlong = pd.merge(table,latlong,how='left', on="Postal Code") ##pd.merge(left, right, how='left', on=['key1', 'key2'])
```

```
tablewithlatlong.shape
```

```
tablewithlatlong.head()
```

Let's see how many rows have no assigned boroughs.

```
noassignedborough= tablewithlatlong[tablewithlatlong.Borough=="Not assigned"]
noassignedborough.shape
```

Now, let's drop those with no assigned boroughs and see the resulting table.

```
boroughassigned = tablewithlatlong.drop(noassignedborough.index)
boroughassigned
```

Let's see how many rows we ended up with

```
boroughassigned.shape
```
